```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/cleaned/telco_cleaned.csv')
print(df.shape)
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Churn')
plt.title('Churn Distribution')
plt.xticks([0, 1], ['No Churn', 'Churned'])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Churn by Contract Type')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(data=df, x='tenure', hue='Churn', bins=30, multiple='stack')
plt.title('Tenure vs Churn')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges')
plt.title('Monthly Charges vs Churn')
plt.xticks([0, 1], ['No Churn', 'Churned'])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
numeric_df = df[['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']]
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

## Key Insights

1. Overall churn rate is 26.58%
2. Month-to-month customers churn at 42% vs 2.85% for two-year contracts
3. Electronic check users churn at 45% — 3x higher than automatic payment methods
4. Nearly 48% of customers leave within first 12 months
5. Fiber optic customers churn at 42% despite being premium service
6. Higher monthly charges correlate with higher churn
